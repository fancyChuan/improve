```python
from pydub import AudioSegment

# 读取音频文件
audio = AudioSegment.from_mp3("mp3/Ahxello-Infinity.mp3")

# 定义要提取的时间段
start_time = 10 * 1000 # 开始时间为10秒
end_time = 20 * 1000 # 结束时间为20秒

# 从音频中提取指定时间段的音频
extracted_audio = audio[start_time:end_time]

# 将提取的音频保存为新的文件
extracted_audio.export("extracted_audio.wav", format="wav")
```

```python
import librosa
y,sr = librosa.load("extracted_audio.wav")
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
print(mfccs)
```
