### networkx
#### 1.1  有向图

```python
import networkx as nx
```

DiGraphe是有向图，通过add_node方法添加节点

```python
g = nx.DiGraph()
g.add_node("Alice")
g.add_node("Bob")
g.add_node("Chuck")
g.nodes()
```

通过add_edge()方法来添加边的关系

```python
g.add_edge("Alice", "Bob")
g.add_edge("Alice", "Chuck")
g.add_edge("Bob", "Alice")
g.add_edge("Bob", "Chuck")
g.edges()
```

networkx提供了多种绘图功能，draw_circular将节点排列成一个圆圈，并将它们的边连起来

```python
nx.draw_circular(g, with_labels=True, node_size=2000)
```

#### 1.2 带权重的无向图

```python
positions = dict(Albany=(-74,43)
                , Boston=(-71,42)
                , NYC=(-74, 41)
                , Philly=(-75, 40))
G = nx.Graph()
G.add_nodes_from(positions)
G.nodes()
```
